### Las columnas de los datos:
- 0 FECHA: fecha de registro
- 1 DESDE: horario de inicio del registro
- 2 HASTA: horario de finalización del registro
- 3 LINEA: linea de subte
- 4 MOLINETE: identificacion del molinete
- 5 ESTACION: estación de subte
- 6 pax_pagos: cantidad de pasajeros que pagaron
- 7 pax_pases_pagos: cantidad de pasajeros que pagaron con pase
- 8 pax_franq: cantidad de pasajeros que pagaron con franquicias
- 9 pax_TOTAL: cantidad total de pasajeros que usaron el molinete

### 1-Calcular la cantidad total de pasajeros que pagaron (pax_pagos) por línea
Sabiendo que tenemos que lineas principales 

```python
subtePaxPagos[0]
```

```python
p = subtePaxPagos
#cantidad de datos por columna
#print(np.unique(p[1:,0], return_counts = True))
print(np.unique(p[1:,1], return_counts = True))
print(p.shape[0])
```

```python
print(p)
#print(p.shape)
#print(p.shape[0])
#print(p.shape[1])
#print(p[0,])
```

```python
lineaD = 0
lineaE = 0
lineaH = 0
otroPago = 0 

for x in range(1,p.shape[0]):
        if p[x,0] != '0':    
                if p[x,1] == 'LineaD':
                        lineaD += 1
                elif p[x,1] == 'LineaE':
                        lineaE += 1
                elif p[x,1] == 'LineaH':
                        lineaH += 1
        else:
                otroPago += 1
print(f'Pagos Linea D {lineaD}')
print(f'Pagos Linea E {lineaE}')
print(f'Pagos Linea H {lineaH}')
print(F'EL total pagados es de: {lineaD+lineaE+lineaH}')
print(f'Otras formas de pagos: {otroPago}')
```

### 2. Calcular cuantos pasajeros pagaron con pase el día 14/8

```python
pf = subtePagoFecha
#print(pf)
pf[1,0]
#print(np.unique(pf[1:,0]))
print(np.unique(pf[1:,0], return_counts = True))
pf.shape
```

```python
#ay = str(input('Ingrese el dia: '))
#onth = str(input('Ingrese el mes'))
#ear = str(input('Ingrese el año: '))
#echa = f'{day}/{month}/{year}'

pasajeros = 0
fecha = '14/8/2023'
for x in range(1,pf.shape[0]):
    if pf[x,0] == fecha:
        pasajeros += 1
print(f'En la fecha {fecha} pagaron {pasajeros} pasajeros')
```
